```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.cos(x / 2) + np.sin( 3 * x / 4) + np.sin(np.random.normal(0, 0.1, 100))
plt.plot(x, y * 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sin(x) + cos(x / 2) + sin(3x / 4) + noise')
```
